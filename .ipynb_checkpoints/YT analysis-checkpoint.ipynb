```python
#function extracting channel details
def get_channel_stats(youtube, channel_ids):
    all_data = []
    #accessing the api by making a request
        #calling list method and what we want
    request = youtube.channels().list(
                part='snippet, contentDetails, statistics',
                #because its a list
                id = ','.join(channel_ids))
                #for each channel id there's another list index
    
    #going to out put a library
    response = request.execute()
    #accesing items in dictionary and storing them in the data dictionary(json makes it cleaner)
        #item is a list and there's only one index
    #so using for loop to loop through indexes
    for i in range(len(response['items'])):   
        data = dict(Channel_name = response['items'][i]['snippet']['title'],
                    Subscribers = response['items'][i]['statistics']['subscriberCount'],
                    Views = response['items'][i]['statistics']['viewCount'],
                    Total_videos = response['items'][i]['statistics']['videoCount'],
                    #getting all of their videos - can find where uploads id is with JSON formatter
                        #can use playlist id to extract all videos of channel
                    playlist_id = response['items'][i]['contentDetails']['relatedPlaylists']['uploads'])
        #storing all of the channels data in this list
        all_data.append(data)
               
    
    
    
    return all_data
    
```

```python
#getting video data of marvel
#playlist id is called "uploads" - check in JSON converter
#function getting video ids from playlist_ids
#using playlistItems(in youtube API) method to do this

def get_video_ids(youtube, playlist_id):
    
    request = youtube.playlistItems().list(
                part='contentDetails',
                playlistId = playlist_id,
                maxResults = 50)
    response = request.execute()
    
    #list containign all of video ids
    video_ids = []
    #extracting all of video ids
    for i in range(len(response['items'])):
        #find where video ids area located with JSON formatter
        video_ids.append(response['items'][i]['contentDetails']['videoId'])
        
    #if there is more than 50 video ids, there is more pages
        #.get (if there is a next page token then get it)
    next_page_token = response.get('nextPageToken')
    #theres more than 1 page
    more_pages = True
    
    while more_pages:
        #there aren't any other pages
        if next_page_token is None:
            more_pages = False
        else:
        #if there's more pages then keep fetching video ids
            request = youtube.playlistItems().list(
                part='contentDetails',
                playlistId = playlist_id,
                #more results per page(want to output all of them)
                maxResults = 50,
                #
                pageToken = next_page_token)
            response = request.execute()
            #getting the remaining 50 video ids
            for i in range(len(response['items'])):
                #find where video ids area located with JSON formatter
                video_ids.append(response['items'][i]['contentDetails']['videoId'])
        
            #check if there's another page to get video ids from
            next_page_token = response.get('nextPageToken')
    
    #should now print all of video ids from MrBeast's channel
    return video_ids
```

```python
video_ids = get_video_ids(youtube, playlist_id)
```

```python
#function to get video details by video methods
def get_video_details(youtube, video_ids):
    
    all_video_stats = []
    #outputting all of video details 50 at a time
    for i in range(0, len(video_ids), 50):
        request = youtube.videos().list(
                #getting snippet and statistics of each video id
                part = 'snippet,statistics',
                #can only pass 50 ids at a time
                id=','.join(video_ids[i:i+50]))
        response = request.execute()
        
        #have to loop through 50 times to get video details
        for video in response['items']:
            #storing all details in this dictionary
                #check where they're located
            video_stats = dict(Title = video['snippet']['title'],
                              Published_date = video['snippet']['publishedAt'],
                              Views = video['statistics']['viewCount'],
                              Likes = video['statistics']['likeCount'],
                              Comments = video['statistics']['commentCount']
                              )
            all_video_stats.append(video_stats)
            
    
    return all_video_stats
```

```python
video_details = get_video_details(youtube, video_ids)
```

```python
video_data = pd.DataFrame(video_details)
#creating data fram to display video details of all videos
```
